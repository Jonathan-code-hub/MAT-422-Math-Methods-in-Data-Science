2.3.1 Joint Probability Distributions

```python
import numpy as np
import pandas as pd

# Define the joint probability distribution table for two discrete random variables X and Y #
joint_prob_table = pd.DataFrame({
    'Y=1': [0.1, 0.2, 0.1],  # Probabilities for Y=1 given different values of X #
    'Y=2': [0.2, 0.1, 0.3],  # Probabilities for Y=2 given different values of X #
    'Y=3': [0.1, 0.0, 0.1]   # Probabilities for Y=3 given different values of X #
}, index=['X=1', 'X=2', 'X=3'])

# Display the joint probability table #
print("Joint Probability Distribution Table:")
print(joint_prob_table)

# Marginal probability of X by summing over all Y values #
marginal_prob_X = joint_prob_table.sum(axis=1)
print("\nMarginal Probability of X:")
print(marginal_prob_X)

# Marginal probability of Y by summing over all X values #
marginal_prob_Y = joint_prob_table.sum(axis=0)
print("\nMarginal Probability of Y:")
print(marginal_prob_Y)
```

2.3.2 Correlation and Dependence

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples for two variables X and Y #
np.random.seed(42)  # For reproducibility #
X = np.random.rand(100)
Y = 0.5 * X + np.random.normal(0, 0.1, 100)  # Y is dependent on X with some added noise #

# Calculate the Pearson correlation coefficient #
correlation = np.corrcoef(X, Y)[0, 1]

print(f"Pearson correlation coefficient between X and Y: {correlation:.2f}")

# Plotting the relationship #
plt.scatter(X, Y, color='blue', alpha=0.5)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

2.3.3 Random Samples

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution #
mean = 0  # Mean ()
std_dev = 1  # Standard deviation () #

# Generate a random sample of size 1000 from the normal distribution #
sample_size = 1000
samples = np.random.normal(mean, std_dev, sample_size)

# Calculate sample statistics #
sample_mean = np.mean(samples)
sample_variance = np.var(samples)

print(f"Sample mean: {sample_mean:.2f}")
print(f"Sample variance: {sample_variance:.2f}")

# Plotting the histogram of the samples #
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# Plot the theoretical normal distribution PDF #
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (np.sqrt(2 * np.pi) * std_dev)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
plt.plot(x, p, 'k', linewidth=2)
title = "Histogram of Samples and Normal Distribution PDF"
plt.title(title)
plt.show()
```

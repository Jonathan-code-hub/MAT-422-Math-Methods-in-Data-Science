Probability Axioms

```python
import numpy as np

outcomes = ['A', 'B', 'C']

# Probabilities assigned to each outcome #
probabilities = np.array([0.2, 0.5, 0.3])

# Axiom 1: Non-negativity #
# None of the Probabilties Should Be Negative #
non_negativity_check = all(p >= 0 for p in probabilities)
print(f"Axiom 1 (Non-negativity): {non_negativity_check}")

# Axiom 2: Normalization #
# The Sum of Probabilties Should Be One #
normalization_check = np.isclose(np.sum(probabilities), 1)
print(f"Axiom 2 (Normalization): {normalization_check}")

# Axiom 3: Additivity (mutually exclusive events) #
# Adding Probilities Should Be the Same as the Union of the Probabilities #
event_A = probabilities[0]  # Probability of event an A #
event_B = probabilities[1]  # Probability of event an B
union_AB = event_A + event_B  # Probability of event A or B

print(f"Axiom 3 (Additivity): P(A ∪ B) = {union_AB} | P(A) + P(B) = {event_A + event_B}")

# Testing the additivity with a third an event C #
event_C = probabilities[2]  # Probability of an event C #
union_ABC = event_A + event_B + event_C  # Probability of A, B, or C #

print(f"Total Probability P(A ∪ B ∪ C) = {union_ABC} | Should be 1: {np.isclose(union_ABC, 1)}")
```

Conditional Probability

```python
import numpy as np

# Define probabilities of events #
P_A = 0.4  # Probability of event A #
P_B = 0.5  # Probability of event B #
P_A_and_B = 0.2  # Probability of both A and B occurring simultaneously #

# Axiom check for probabilities #
# Check if the probability of event B is zero to avoid division by zero in conditional probability #
if P_B == 0:
    print("Error: P(B) cannot be zero for conditional probability.")
else:
    # Calculate conditional probability P(A | B) #
    P_A_given_B = P_A_and_B / P_B

    print(f"P(A) = {P_A}")
    print(f"P(B) = {P_B}")
    print(f"P(A ∩ B) = {P_A_and_B}")
    print(f"P(A | B) = {P_A_given_B:.2f}")
```

Discrete Random Variables

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the discrete random variable: outcomes of rolling a die #
outcomes = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6] * 6)  # Uniform distribution for a fair die #

# Calculate the expected value (mean) #
expected_value = np.sum(outcomes * probabilities)
print(f"Expected value (E[X]) of rolling a die: {expected_value:.2f}")

# Calculate the variance #
variance = np.sum(((outcomes - expected_value) ** 2) * probabilities)
print(f"Variance (Var[X]) of rolling a die: {variance:.2f}")

# Plotting the PMF #
plt.bar(outcomes, probabilities, color='skyblue', alpha=0.7)
plt.xticks(outcomes)
plt.title("Probability Mass Function of Rolling a Six-Sided Die")
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.ylim(0, 0.2)
plt.grid(axis='y')
plt.show()
```

Continous Random Variables

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution #
mean = 0      # Mean (μ) #
std_dev = 1   # Standard deviation (σ) #

# Generate a range of x values #
x = np.linspace(-4, 4, 1000)
# Calculate the probability density function (PDF) #
pdf = norm.pdf(x, mean, std_dev)

# Plotting the PDF #
plt.plot(x, pdf, color='skyblue', label='Normal Distribution')
plt.fill_between(x, pdf, alpha=0.2)
plt.title("Probability Density Function of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label='Mean (μ)')
plt.legend()
plt.grid()
plt.show()

# Calculate probabilities for specific intervals #
P_between_neg1_and_1 = norm.cdf(1, mean, std_dev) - norm.cdf(-1, mean, std_dev)
print(f"Probability that X is between -1 and 1: {P_between_neg1_and_1:.4f}")

# Calculate expected value (mean) and variance #
expected_value = mean
variance = std_dev ** 2
print(f"Expected value (E[X]): {expected_value}")
print(f"Variance (Var[X]): {variance:.4f}")
```
